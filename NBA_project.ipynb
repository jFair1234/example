# Is defense getting worse or has the nba changed how its played to cause this spike in PPG.

The first thing i did was compare the average Defense from the nba in recent years (2010 - 2023) to the average defense in past years (1947 - 2010).

```python
year_cut = pd.cut(D_RTG['season'], [1947, 2010, 2023], labels = ['Past Years', 'Recent Years'])
```

NBA defensive rating is a statistic that measures how many points a player or a team allows per 100 possessions. So the lower the number the bwtter the rating is.

```python
 x = D_RTG.groupby(year_cut)['d_rtg'].mean()
 x
```

```python
fig, ax = plt.subplots()
#5 wide 15 tall
ax.bar(x.index, x)

plt.title('Average Defensive Rating')

ax.set(xlabel = 'Age', ylabel = 'Defensive rating')


plt.yticks(range(0, 115, 10))
plt.show()
```

After comparing the 2 on a pivot table and visibly seein git on a graph you can see how deffense rating on average was better before in past years  than it is now in recent years. So by statistics deffense was better before than it is now.

Though, is that deffense rating a direct relation to the PPG spike.

```python
year_cut_ppg = pd.cut(T_PPG['season'], [1947, 2010, 2023], labels = ['Past Years', 'Recent Years'])
```

```python
 y = T_PPG.groupby(year_cut_ppg)['pts_per_game'].mean()
 y
```

```python
fig, ax = plt.subplots()

ax.bar(y.index, y)

plt.title('Average Points per Game')

ax.set(xlabel = 'Age', ylabel = 'PPG')

plt.yticks(range(0, 115, 10))
plt.show()
```

So as you can see the average defense rating does have some influence on the points per game these erras have been averaging. But, this got me to thinking. Is this increase in points per game just now happening or has it been increasing for these past years and we are just now noticeing.

```python
const_year_cut= pd.cut(T_PPG['season'], [1947, 1957, 1967, 1977, 1987, 1997, 2007, 2017, 2023], labels = ['1947 - 1957',
                                                                                                          '1957 - 1967',
                                                                                                          '1967 - 1977',
                                                                                                          '1977 - 1987',
                                                                                                          '1987 - 1997',
                                                                                                          '1997 - 2007',
                                                                                                          '2007 - 2017',
                                                                                                          '2017 - 2023'])
```
